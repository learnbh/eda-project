```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_detail = pd.read_csv('data/clean_king_county_house_details.csv', sep=';')
df_sales = pd.read_csv('data/clean_king_county_house_sales.csv', sep=';')
```

```python
# Floats (decimal numbers) should be displayed rounded with 2 decimal places
pd.options.display.float_format = "{:,.2f}".format
```

```python
print(df_detail.shape)
print(df_sales.shape)
```

```python
list_houseIds_with_waterfront = df_detail.query("waterfront == 'yes'")['id'].to_list()
list_houseIds_with_waterfront
```

```python
df_detail.query("waterfront == 'yes'")
```

```python
df_detail.query("waterfront == 'no'")
```

```python

list_houseIds_without_waterfront = df_detail.query("waterfront == 'no'")['id'].to_list()
list_houseIds_without_waterfront
```

```python
df_prices_waterfront = df_sales.query("house_id in @list_houseIds_with_waterfront")\
    .sort_values(["house_id", "date"], ascending=[True, False])\
    .drop_duplicates(subset="house_id", keep="first")\
    .reset_index(drop=True)   
df_prices_waterfront
df_prices_waterfront.house_id.duplicated().value_counts()
df_prices_waterfront
```

```python
df_prices_no_waterfront = df_sales.query("house_id in @list_houseIds_without_waterfront").sort_values(["house_id", "date"], ascending=[True, False]).drop_duplicates(subset="house_id", keep="first").reset_index(drop=True)   
df_prices_no_waterfront
df_prices_no_waterfront.house_id.duplicated().value_counts()
```

```python
# only the last price is important for the analysis
df_all_prices = df_sales\
    .sort_values(["house_id", "date"], ascending=[True, False])\
    .drop_duplicates(subset="house_id", keep="first")\
    .reset_index(drop=True)
df_all_prices
```

```python
fig, ax = plt.subplots(2,3, figsize=(10,10))#, sharey=True) # create subplots on 2 rows and 3 columns
plt.suptitle('Comparing price of waterhouses with no waterhouses', fontsize=20)
fig.tight_layout() # tight_layout automatically adjusts subplot params so that the subplot(s) fits in to the figure area
plt.subplots_adjust(hspace = .5, wspace = .2, top = .9) # adjusts the space between the single subplots

ax[0][0].boxplot(x = df_prices_waterfront['price']) # creates boxplot for price
ax[0][0].set_xticklabels(labels=["Price"]) # sets the label for the ticks on the x-axis
ax[0][0].set_ylim(0,8000000) # limits the y-axis values from 0 to 2000000 
ax[0][0].set_title("prices for waterfront", fontsize = 15); # sets title for subplot

ax[0][1].boxplot(x = df_prices_no_waterfront['price'])
ax[0][1].set_xticklabels(labels=["Price"])
ax[0][1].set_ylim(0,8000000)
ax[0][1].set_title("prices for no waterfront", fontsize = 15);

ax[0][2].boxplot(x = df_all_prices['price'])
ax[0][2].set_xticklabels(labels=["Price"])
ax[0][2].set_ylim(0,8000000)
ax[0][2].set_title("prices in of all sales", fontsize = 15);

fig.delaxes(ax[1][0]);
fig.delaxes(ax[1][1]);
fig.delaxes(ax[1][2]);
```

Comparing the values, then the data between min and max of the 2. box are in the lower part, also under the median of the distribution with waterhouse.
But there are also a lot of outliners with high prices of houses with no waterfront - beginning with max of no waterfront to max of waterhouses. So, there might be other variables for high prices?
Also, to consider there are just 146 waterhouses saled. All the other houses (18914) are without waterfront. 

```python
df_details_prices = df_detail.copy()
df_details_prices = df_details_prices.merge(
        df_all_prices[['house_id', 'price']],      # nur die Spalten, die wir brauchen
        left_on='id',                               # Schlüssel in dfDP
        right_on='house_id',                        # Schlüssel in dfAP
        how='left'                                  # „left“ ⇒ alles aus dfDP bleibt erhalten
    )
df_details_prices.rename(columns={'price': 'last_sale_price'}, inplace=True)
df_details_prices.drop(columns=['house_id'], inplace=True)
df_details_prices
```

```python
df_details_prices_no_waterhouse = df_details_prices.query("waterfront == 'no'")
df_details_prices_waterhouse = df_details_prices.query("waterfront == 'yes'")
fig, ax_sns = plt.subplots(2,1, figsize=(10, 10),
    sharex=True      )
plt.suptitle('Distribution for prices waterhouse/no waterhouse', fontsize=20) 
#sns.set_theme(style="white", palette="Blues")
my_palette = sns.color_palette([ "#17BECF"], n_colors=1)

f1 = sns.histplot(data=df_details_prices_no_waterhouse,
             x='last_sale_price',
             hue = 'waterfront',
             ax=ax_sns[0], 
             palette=my_palette 
)
f1.set_xlabel("Last saled price", fontsize = 10)
f1.set_ylabel("Count of houses", fontsize = 10);  

f2= sns.histplot(data=df_details_prices_waterhouse,
             x='last_sale_price',
             hue = 'waterfront',
             ax=ax_sns[1],
             palette=my_palette 
)
f2.set_xlabel("Last saled price", fontsize = 10)
f2.set_ylabel("Count of houses", fontsize = 10);

plt.show()
```

```python
df_details_prices_no_nan_warehouse = df_details_prices.query("waterfront != 2")
fig, ax_sns = plt.subplots(figsize=(14, 6))
sns.set_theme(style="white", palette="Blues")
my_palette = sns.color_palette([ "#8c4b6f","#101111" ,"#17BECF" ], n_colors=3)
sns.histplot(data=df_details_prices,
             x='sqft_living',
             y = 'last_sale_price',
             hue = 'waterfront',
             ax=ax_sns,
             palette=my_palette
)
ax_sns.set_xlabel(xlabel='Last saled price', fontsize=10)
ax_sns.set_ylabel(ylabel='Sqft for living', fontsize=14)
ax_sns.set_title(label="Infects a higher living sqft the price for waterhouses", fontsize=16) 
plt.show()
```

```python
# Plot views, likes, dislikes and comments per category
fig, ax = plt.subplots(2,2, figsize=(15,10)) # create subplots on 2 rows and 2 columns
plt.suptitle('sqft_living, condition, views and grade in corrolation to price waterhouse/no waterhouse', fontsize=20) 
fig.tight_layout()   # tight_layout automatically adjusts subplot params so that the subplot(s) fits in to the figure area
plt.subplots_adjust(hspace = .6, wspace = .2, top = .9) # adjusts the space between the single subplots

f1 = sns.histplot(data=df_details_prices_no_nan_warehouse,
             x='sqft_living',
             y = 'last_sale_price',
             hue = 'waterfront',
             ax=ax[0,0],
             palette=my_palette) # creates barchart for number of views
f1.set_ylabel("Last saled price", fontsize = 10)# sets the x-axis label
f1.set_xlabel("Sqft for living", fontsize = 10) # limits the x-axis values from 0 to 1000000
f1.set_title("Price per sqft for living", fontsize = 15);  # sets title for subplot

f2= sns.histplot(data=df_details_prices_no_nan_warehouse,
             x='view',
             y = 'last_sale_price',
             hue = 'waterfront',
             ax=ax[0,1],
             palette=my_palette) 
f2.set_ylabel("Last saled price", fontsize = 10)
f2.set_xlabel("View", fontsize = 10)
f2.set_title("Price per Views", fontsize = 15)

f3= sns.histplot(data=df_details_prices_no_nan_warehouse,
             x='condition',
             y = 'last_sale_price',
             hue = 'waterfront',
             ax=ax[1,0],
             palette=my_palette) 
f3.set_ylabel("Last saled price", fontsize = 10)
f3.set_xlabel("Condition", fontsize = 10)
f3.set_title("Price per Conditions", fontsize = 15)

f3= sns.histplot(data=df_details_prices_no_nan_warehouse,
             x='grade',
             y = 'last_sale_price',
             hue = 'waterfront',
             ax=ax[1,1],
             palette=my_palette) 
f3.set_ylabel("Last saled price", fontsize = 10)
f3.set_xlabel("Grade", fontsize = 10)
f3.set_title("Price per Grade", fontsize = 15);
```
