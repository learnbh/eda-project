```python
df_details_copy.yr_renovated.isna().sum()
```

```python
df_details_copy.yr_renovated = df_details_copy.fillna({'yr_renovated':5})['yr_renovated']
```

```python
df_details_copy['yr_renovated'] = df_details_copy.yr_renovated.apply(lambda x: x/10 if (x > 10000) else x )
df_details_copy.yr_renovated.loc[13]
```

```python
df_details_copy.yr_renovated.dtype
```

```python
df_details_copy['yr_renovated'] = df_details_copy.yr_renovated.astype('int')
df_details_copy.yr_renovated.dtype
```

```python
df_details_copy.yr_renovated.max()
```

```python
# Lesbare Kategorien (z. B. Plot‑Legenden)
df_details_clean_data = df_details_copy.copy()
df_details_clean_data['yr_renovated_cat'] = df_details_copy.yr_renovated.astype('str')
df_details_clean_data['yr_renovated_cat'].dtype
```

```python

df_details_clean_data['yr_renovated_cat'] = df_details_clean_data.yr_renovated_cat.map(lambda x: x.replace('5', 'unkown') if (x == '5') else x )
df_details_clean_data['yr_renovated_cat'] = df_details_clean_data.yr_renovated_cat.map(lambda x: x.replace('0', 'no renovation') if (x == '0') else x )
```

```python
df_details_clean_data['yr_renovated_cat'] = df_details_copy.yr_renovated.astype('category')
df_details_clean_data.head(20)
```

```python
# waterfront should be boole not float, or better int ( 0: false, 1 = true, 5 = unknown ) -> category
df_details_clean_data.waterfront = df_details_clean_data.fillna({'waterfront':5})['waterfront'] # replace of NaNs with 5
```

```python
df_details_clean_data["waterfront"] = df_details_clean_data["waterfront"].astype("category")
# Optional: Kategorienamen direkt beim Casten setzen
df_details_clean_data["waterfront"] = df_details_clean_data["waterfront"].cat.rename_categories({0: "no", 1: "yes", 5:"unkown"})
df_details_clean_data.head(20)
```

```python
df_details_clean_data.waterfront.head()
```

```python
# 2. view should be int -> is a category 
# first step get rid of Nans -> 5 = unkown
df_details_clean_data.view.isna().value_counts()
df_details_clean_data.view.describe()
df_details_clean_data.view = df_details_clean_data.fillna({'view':5})['view']
df_details_clean_data.view = df_details_clean_data.view.astype('int')
# set 
df_details_clean_data["view"] = df_details_clean_data["view"].astype("category")
# Optional: Kategorienamen direkt beim Casten setzen
df_details_clean_data["view"] = df_details_clean_data["view"].cat.rename_categories({5:"unkown"})
```

```python
# 3. condition should be int -> is a category 
# first step get rid of Nans -> 5 = unkown
df_details_clean_data.condition.isna().value_counts()
df_details_clean_data.condition.describe()
# set 
df_details_clean_data["condition"] = df_details_clean_data["condition"].astype("category")
```

```python
# 4. grade should be int -> is a category 
# first step get rid of Nans -> 5 = unkown
df_details_clean_data.condition.isna().value_counts()
df_details_clean_data.condition.describe()
# set 
df_details_clean_data["grade"] = df_details_clean_data["grade"].astype("category")
```
